```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Model performance data
data = {
    "Model": [
        "XGBoost",
        "Random Forest A",
        "Random Forest B"
    ],
    "Accuracy": [0.978, 0.89, 0.974],
    "Recall (Churn)": [0.91, 0.92, 0.89],
    "F1-score": [0.91, 0.89, 0.92],
    "CV Accuracy": [0.982, 0.89, "—"]
}

df = pd.DataFrame(data)
df_sorted = df.sort_values("Recall (Churn)", ascending=False)
df_sorted
```

```python
# Plot comparison
ax = df_sorted.plot(
    x="Model",
    y=["Accuracy", "Recall (Churn)", "F1-score"],
    kind="bar",
    figsize=(10, 6)
)

plt.title("Model Performance Comparison")
plt.ylabel("Score")
plt.ylim(0.7, 1.0)
plt.grid(True)
plt.xticks(rotation=0)
plt.legend(loc="lower right")
plt.tight_layout()
plt.show()
```

Based on recall and cross-validation accuracy, XGBoost outperforms both Random Forest models in identifying churners. While Random Forest A offers strong recall and stakeholder-friendly framing, and Random Forest B shows high precision and F1-score, XGBoost balances generalization and early detection, making it the best candidate for deployment.
